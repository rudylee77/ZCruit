```python
import re

vals = data["Height"].values
for element in vals:
    z = re.match("^(\d)'\s*(\d*\.*\d*)", element)
    if z:
        print((z.groups()))
```

```python
height_regex = "^(\d)'\s*(\d*\.*\d*)"
def height_str_to_val(height_str):
    match = re.match("^(\d)'\s*(\d*\.*\d*)", height_str)
    groups = match.groups()
    feet, inches = int(groups[0]), float(groups[1])
    return feet * 12 + inches
```

```python
data['Height'] = data['Height'].apply(height_str_to_val)
```

```python
def create_percentile(position, event):
    """Take in a position - CB, WR, QB, etc. and an event - Fourty Yard Dash, Weight, 3 Cone and return a list of percentile values
        of 0, 10th, ... 100th percentile. Reverse for all speed events (lower is better) such as 40, 3 Cone, and Shuttle."""
    sorted_athletes = data[data['Position'] == position]     
    event_values = (sorted_athletes[event].astype(float)).dropna()
    percentile = []
    for i in np.arange(0, 1.01, 0.01):
        percentile.append(event_values.quantile(i))
    if (event == "Forty Yard Dash") or (event == "Shuttle") or (event == "3 Cone"):
        return percentile[::-1]
    return percentile
```
